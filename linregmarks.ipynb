```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
```

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported successfully")

data.head(10)
```

```python
data.plot(x="Hours",y="Scores",style="o")
plt.show()
```

```python
import seaborn as sns
sns.lineplot(data=data)
```

```python

df0=data[data['Hours']<=5]
plt.title('Students who study less than 5 hours mark in green color')
plt.scatter(df0['Hours'],df0['Scores'],color="green")
df1=data[data['Hours']>5]
plt.scatter(df1['Hours'],df1['Scores'],color="red")
```

```python
plt.bar(data.Hours,data.Scores,color="red")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
x = data[['Hours']]  
y = data['Scores']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
from sklearn.svm import SVC
model = SVC()  
model.fit(x_train, y_train) 

print("Training complete.")
```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
# Comparing Actual vs Predicted
x_test=x_test['Hours']
df = pd.DataFrame({'Hours':x_test,'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
hours=9.5
own_pred = model.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

```python
plt.scatter(x_test,y_pred,color="red")
plt.scatter(x_test,y_test)
plt.title("Actual vs Predict")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```
