```python
#Data analysis
new_state=data[data['Profit']<0]
print(new_state)
```

# Data analysis on which factors most effect on loss in company?

```python
#States which is in loss
#so we should work on these states
```

```python
m=new_state.groupby('State')[['Profit','Sales']].sum().sort_values(by="Sales")
m.plot.bar(color=["red","blue"],figsize=(16,8))
```

```python
#Office supplies and Furniture are in loss
#So we should work on two
```

```python
m=new_state.groupby('Category')[['Profit','Sales']].sum().sort_values(by="Sales")
m.plot.bar(color=["red","blue"],figsize=(16,8))
```

```python
#you can observe that our loss is most on-Machines,Tables and chairs
```

```python
m=new_state.groupby('Sub-Category')[['Profit','Sales']].sum().sort_values(by="Sales")
m.plot.bar(color=["red","blue"],figsize=(16,8))
```

```python
#IF we start to give less amount of discount then we can reduce the loss
```
